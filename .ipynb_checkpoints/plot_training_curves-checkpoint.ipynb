```python
exp_name = 'rob_wiki_class'
exp_dir = f'/home/alta/Conversational/OET/al826/2021/coherence/results/synthetic/{exp_name}'
temp_dir = f'/home/alta/Conversational/OET/al826/2021/coherence/results/temp'

log_path = f'{exp_dir}/train_curves.json'
```

```python
import json 

with open(log_path, 'r') as f:
    data = json.load(f)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme()
train_loss = data['train_loss']
train_acc = data['train_acc']

x_train = 100*np.arange(1, len(train_acc)+1)
plt.plot(x_train, train_loss)
plt.plot(x_train, train_acc)


mean = data['gcdc_mean']
spearman = data['gcdc_spear']
dev_acc = data['dev_acc']

x_dev = 500*np.arange(1, len(spearman)+1)
plt.plot(x_dev, spearman)
plt.plot(x_dev, dev_acc)
```
